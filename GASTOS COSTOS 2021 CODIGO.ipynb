```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
datos_gastos_costos_2021 = pd.read_csv('GASTOS Y COSTOS 2021 CORREGIDO (1).csv')
datos_gastos_costos_2021
```

```python
datos_gastos_costos_2021.info()
```

```python
columnas_cualitativas = datos_gastos_costos_2021[['Unnamed: 0', 'FECHA', 'FOLIO', 'UUID', 'RFC', 'PROVEEDOR', 'TIPO GASTO',  'DESCRIPCION', 'MP', 'TIPO', 'STATUS', 'POLIZA']]
```

```python
datos_gastos_costos_2021_1 = datos_gastos_costos_2021.drop(['Unnamed: 0', 'FECHA', 'FOLIO', 'UUID', 'RFC', 'PROVEEDOR', 'TIPO GASTO',  'DESCRIPCION', 'MP', 'TIPO', 'STATUS', 'POLIZA'], axis = 1)
datos_gastos_costos_2021_1
```

```python
#Realizamos diagrama de caja o bigote de cada columna del data frame
fig = plt.figure(figsize = (15, 8))
datos_gastos_costos_2021_1.plot(kind = 'box', vert = False)
plt.title("Valores atipicos del DataFrame")
plt.show()
```

```python
y = datos_gastos_costos_2021_1
Limite_superior = y.mean() + 3*y.std()
Limite_inferior = y.mean() - 3*y.std()
print("Limite superior permitido", Limite_superior)
print("Limite inferior permitido", Limite_inferior)
```

```python
gastos_costos_2021_limpio = datos_gastos_costos_2021_1[(y <= Limite_superior) & (y >= Limite_inferior)]
gastos_costos_2021_limpio
```

```python
#Corroboramos valores nulos del dataframe
valores_nulos = gastos_costos_2021_limpio.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores atipicos (nulos) del dataframe con "mean"
#Realizamos una copia del dataframe
gastos_costos_2021_limpioo = gastos_costos_2021_limpio.copy()
gastos_costos_2021_limpioo = gastos_costos_2021_limpioo.fillna(round(gastos_costos_2021_limpio.mean(), 1))
gastos_costos_2021_limpioo
```

```python
valores_nulos_data_clean = gastos_costos_2021_limpioo.isnull().sum()
valores_nulos_data_clean
```

```python
gastos_costos_2021_final = pd.concat([gastos_costos_2021_limpioo, columnas_cualitativas], axis=1)
gastos_costos_2021_final
```

```python
gastos_costos_2021_final.to_csv("Gastos y costos 2021 Limpio.csv")
```
