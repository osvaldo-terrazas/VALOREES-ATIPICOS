```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
datos_gastos_costos_2023 = pd.read_csv('GASTOS Y COSTOS 2023 CORREGIDO.csv')
datos_gastos_costos_2023
```

```python
datos_gastos_costos_2023.info()
```

```python
datos_gastos_costos_2023_1 = datos_gastos_costos_2023.drop(['Unnamed: 0', 'Fecha', 'Folio', 'UUID', 'RFC', 'Proveedor', 'TIPO GASTO',  'Descripción', 'MP', 'Tipo', 'Status', 'Poliza', 'Otros'], axis = 1)
datos_gastos_costos_2023_1
```

```python
fig = plt.figure(figsize = (15, 8))
datos_gastos_costos_2023_1.plot(kind = 'box', vert = False)
plt.title("Valores atipicos del DataFrame")
plt.show()
```

```python
y = datos_gastos_costos_2023_1
Limite_superior = y.mean() + 3*y.std()
Limite_inferior = y.mean() - 3*y.std()
print("Limite superior permitido", Limite_superior)
print("Limite inferior permitido", Limite_inferior)
```

```python
gastos_costos_2023_limpio = datos_gastos_costos_2023_1[(y <= Limite_superior) & (y >= Limite_inferior)]
gastos_costos_2023_limpio
```

```python
#Corroboramos valores nulos del dataframe
valores_nulos = gastos_costos_2023_limpio.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores atipicos (nulos) del dataframe con "mean"
#Realizamos una copia del dataframe
gastos_costos_2023_limpioo = gastos_costos_2023_limpio.copy()
gastos_costos_2023_limpioo = gastos_costos_2023_limpioo.fillna(round(gastos_costos_2023_limpio.mean(), 1))
gastos_costos_2023_limpioo
```

```python
valores_nulos_data_clean = gastos_costos_2023_limpioo.isnull().sum()
valores_nulos_data_clean
```

```python
gastos_costos_2023_limpioo.to_csv("Gastos y costos 2023 Limpio.csv")
```
