```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
datos_gastos_costos_2020 = pd.read_csv('GASTOS Y COSTOS 2020 CORREGIDO.csv')
datos_gastos_costos_2020
```

```python
datos_gastos_costos_2020.info()
```

```python
columnas_cualitativas = datos_gastos_costos_2020[['Unnamed: 0', 'FECHA', 'FOLIO', 'UUID', 'RFC', 'PROVEEDOR', 'TIPO GASTO', 'GASTO', 'DESCRIPCION', 'TC', 'TIPO', 'STATUS', 'POLIZA', 'Columna1']]
```

```python
#Eliminamos columnas innecesarias
#Eliminamos columnas innecesarias
datos_gastos_costos_2020_1 = datos_gastos_costos_2020.drop(['Unnamed: 0', 'FECHA', 'FOLIO', 'UUID', 'RFC', 'PROVEEDOR', 'TIPO GASTO', 'GASTO', 'DESCRIPCION', 'TC', 'TIPO', 'STATUS', 'POLIZA', 'Columna1'], axis = 1)
datos_gastos_costos_2020_1
```

```python
#Realizamos diagrama de caja o bigote de cada columna del data frame
fig = plt.figure(figsize = (15, 8))
datos_gastos_costos_2020_1.plot(kind = 'box', vert = False)
plt.title("Valores atipicos del DataFrame")
plt.show()
```

```python
y = datos_gastos_costos_2020_1
Limite_superior = y.mean() + 3*y.std()
Limite_inferior = y.mean() - 3*y.std()
print("Limite superior permitido", Limite_superior)
print("Limite inferior permitido", Limite_inferior)
```

```python
gastos_costos_2020_limpio = datos_gastos_costos_2020_1[(y <= Limite_superior) & (y >= Limite_inferior)]
gastos_costos_2020_limpio
```

```python
#Corroboramos valores nulos del dataframe
valores_nulos = gastos_costos_2020_limpio.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores atipicos (nulos) del dataframe con "mean"
#Realizamos una copia del dataframe
gastos_costos_2020_limpioo = gastos_costos_2020_limpio.copy()
gastos_costos_2020_limpioo = gastos_costos_2020_limpioo.fillna(round(gastos_costos_2020_limpio.mean(), 1))
gastos_costos_2020_limpioo
```

```python
valores_nulos_data_clean = gastos_costos_2020_limpioo.isnull().sum()
valores_nulos_data_clean
```

```python
gastos_costos_2020_final = pd.concat([gastos_costos_2020_limpioo, columnas_cualitativas], axis=1)
gastos_costos_2020_final
```

```python
gastos_costos_2020_final.to_csv("Gastos y costos 2020 Limpio.csv")
```
